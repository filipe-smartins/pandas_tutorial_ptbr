# Carregar dados, definir tipos, reordenar e alterar nomes das colunas

```python
datatypes = {
    "Seguradora": str,
    "Valor Solicitado": float,
    "Valor Aprovado": float,
    "Data Decisão": 'datetime64[ns]',
    "Cnpj do Buyer": str,
    "Segmento do Buyer": str,
    "Estado do Buyer": str
}

df = pd.read_excel(
    'dados.xlsx',
    engine='openpyxl',
    usecols=datatypes.keys(),
    dtype=datatypes
)

print("\nShape: ", df.shape, "\n")
df.head()
```

```python
nomes_colunas = {
    "Seguradora": "seguradora",
    "Valor Solicitado": "valor_solicitado",
    "Valor Aprovado": "valor_aprovado",
    "Data Decisão": "data_decisao",
    "Cnpj do Buyer": "buyer_cnpj",
    "Segmento do Buyer": "buyer_segmento",
    "Estado do Buyer": "buyer_estado",
}

df = df.rename(columns=nomes_colunas)

nova_ordem_colunas = nomes_colunas.values()
df = df[nova_ordem_colunas]  # Reordenar as colunas

df.head()
```

```python
df['valor_solicitado'] = df['valor_solicitado'].astype(float)
```

# Tratar valores nulos (na, nan)

[Voltar para o Sumário](#sumario)

```python
df.isna().sum()
```

```python
df[df['buyer_cnpj'].isna()]  # Filtrando dados nulos
```

```python
print("\nShape: ", df.shape, "\n")

df.dropna(subset='buyer_cnpj', inplace=True)  # excluindo dados nulos

print("\nShape: ", df.shape, "\n")
```

```python
df[df['buyer_cnpj'].isna()]  # Filtrando dados nulos
```

```python
df[df['valor_solicitado'].isna()]  # Filtrando dados nulos
```

```python
# Preenchendo dados vazios com a média da coluna

media_valor_solicitado = df['valor_solicitado'].mean()
media_valor_aprovado = df['valor_aprovado'].mean()
media_data_decisao = df['data_decisao'].mean()

df['valor_solicitado'] = df['valor_solicitado'].fillna(media_valor_solicitado)
df['valor_aprovado'] = df['valor_aprovado'].fillna(media_valor_aprovado)
df['data_decisao'] = df['data_decisao'].fillna(media_data_decisao)
```
